# Loading Phase Diagram calculations from pickles and plotting using PetThermoTools

Phase diagram calculations can (depending on the computer and resolution) take a lot of time. Rather than rerunning every calculation it can be useful to simply reload exisitng model results and create a plot from this data. Here we demonstrate how to import a prior phase diagram calculation saved as a pickle and display the phase diagram.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import PetThermoTools as ptt
import pickle
import os
import sys

# this can be useful for visualizing the phase diagrams - hover over the plots to see the phase assemblage
%matplotlib widget

# this allows figures to be saved as svg files with the text preserved
plt.rcParams['svg.fonttype'] = 'none'

# check the version
ptt.__version__
```

For this example we load in the calculation performed using Holland et al. (2018) thermodynamic model as this cannot currently be run using the most recent versions of MAGEMin_C (v1.6.9 and later) as it was replaced by the Green et al. model.

```python
with open("Holland_16wtMgO_Ariskin.pkl", "rb") as f:
    Holland = pickle.load(f)
```

```python
ptt.plot_phaseDiagram(Combined=Holland)
```
